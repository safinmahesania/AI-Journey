### 2D Vectors

```python
v = np.array([3, 3])
print(v)

# Visualize
plt.quiver(0, 0, v[0], v[1], scale=1, scale_units='xy', angles='xy',color='g')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.axhline(0)
plt.axvline(0)
plt.grid(True)
plt.title("2D Vector")
plt.show()
```

### 3D Vectors

```python
from mpl_toolkits.mplot3d import axes3d

v_3d = np.array([2,3,5])
print(v_3d)

# Visualize
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])

ax.quiver(0, 0, 0, v_3d[0], v_3d[1], v_3d[2])
ax.set_title("3D Vector")
# ax.view_init()
plt.show()
```

### Addition of 2D Vectors

```python
v1 = np.array([2, 1])
v2 = np.array([1, 3])
v_sum = v1 + v2

print(v_sum)

# Visualize
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, label='v1',color='blue')
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, label='v2',color='green')
plt.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='red', label='v1 + v2')

plt.xlim(0, 5)
plt.ylim(0, 5)
plt.grid()
plt.legend()
plt.title("Vector Addition")
plt.show()
```

### 2D Vector Scalar Multiplication (Scaling)

```python
v = np.array([2, 1])
scaled_v = 3 * v

print(scaled_v)

# Visualize
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, label='Original',color='green')
plt.quiver(0, 0, scaled_v[0], scaled_v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Scaled')

plt.xlim(0, 8)
plt.ylim(0, 5)
plt.grid()
plt.legend()
plt.title("Vector Scaling")
plt.show()
```

### Linear Combination
Scale some vectors and add them together.

```python
v1 = np.array([1, 2])
v2 = np.array([2, 2])

a = 2
b = 3

linear_combo = a * v1 + b * v2
print(linear_combo)

# Visualize
plt.quiver(0, 0, v1[0], v1[1], scale=1, scale_units='xy', angles='xy', label='v1',color='blue')
plt.quiver(0, 0, v2[0], v2[1], scale=1, scale_units='xy', angles='xy', label='v2',color='green')

plt.quiver(0, 0, linear_combo[0], linear_combo[1],
           scale=1, scale_units='xy', angles='xy', color='red', label='a*v1 + b*v2')

plt.xlim(0, 10)
plt.ylim(0, 12)
plt.grid()
plt.legend()
plt.title("Linear Combination")
plt.show()
```

### Span
All possible linear combinations of those vectors.

#### Visualize span of one vector (a line)

```python
v = np.array([1, 2])

scalars = np.linspace(-5, 5, 20)
points = np.array([s * v for s in scalars])

plt.plot(points[:, 0], points[:, 1], '.')
plt.axhline(0)
plt.axvline(0)
plt.grid()
plt.title("Span of a Single Vector")
plt.show()
```

#### Visualize span of two vectors (plane)

```python
v1 = np.array([1, 0])
v2 = np.array([0, 1])

a_vals = np.linspace(-3, 3, 10)
b_vals = np.linspace(-3, 3, 10)

points = []
for a in a_vals:
    for b in b_vals:
        points.append(a * v1 + b * v2)

points = np.array(points)

plt.scatter(points[:, 0], points[:, 1],marker='.',color='red')
plt.grid()
plt.title("Span of v1 and v2 (Entire Plane)")
plt.show()
```

### Basis Vectors
The smallest set of vectors that can generate the entire space.

```python
e1 = np.array([1, 0])
e2 = np.array([0, 1])

v = np.array([4, -2])

# Represented using basis vectors
v = 4 * e1 + (-2) * e2

plt.quiver(0, 0, e1[0], e1[1], scale=1, scale_units='xy', angles='xy', label='e1',color='blue')
plt.quiver(0, 0, e2[0], e2[1], scale=1, scale_units='xy', angles='xy', label='e2',color='green')
plt.quiver(0, 0, v[0], v[1], scale=1, scale_units='xy', angles='xy', label='e2',color='red')

plt.xlim(-3, 5)
plt.ylim(-3, 5)
plt.grid()
plt.legend()
plt.title("Standard Basis Vectors")
plt.show()
```

### Identity Transformation
Do nothing

```python
A = np.array([
    [1, 0],
    [0, 1]
])

v = np.array([2, 3])
# x = Av
x = A @ v
print(x)
```

### Scaling Transformation

```python
A = np.array([
    [2, 0],
    [0, 3]
])

v = np.array([1, 1])

x = A @ v
print(x)
```

### Rotation Transformation (90 degrees)

```python
A = np.array([
    [0, -1],
    [1,  0]
])
v = np.array([2, 1])

x = A @ v
print(x)
```

### Visualizing a Transformation

```python
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

points = np.array([[i, j] for i in x for j in y])

A = np.array([
    [2, 1],
    [0, 1]
])

transformed = points @ A.T

# Plotting before and after

plt.scatter(points[:, 0], points[:, 1], alpha=0.4,color='blue')
plt.scatter(transformed[:, 0], transformed[:, 1], alpha=0.4,color='red')


plt.grid()
plt.title("Linear Transformation")
plt.show()
```
