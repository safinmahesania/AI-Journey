### Addition of 2D Vectors

```python
v1 = np.array([2, 1])
v2 = np.array([1, 3])
v_sum = v1 + v2

print(v_sum)

# Visualize
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, label='v1', color='blue')
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, label='v2', color='green')
plt.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='red', label='v1 + v2')

plt.xlim(0, 5)
plt.ylim(0, 5)
plt.grid()
plt.legend()
plt.title("Vector Addition")
plt.show()
```

### Linear Combination
Scale some vectors and add them together.

```python
v1 = np.array([1, 2])
v2 = np.array([2, 2])

a = 2
b = 3

linear_combo = a * v1 + b * v2
print(linear_combo)

# Visualize
plt.quiver(0, 0, v1[0], v1[1], scale=1, scale_units='xy', angles='xy', label='v1', color='blue')
plt.quiver(0, 0, v2[0], v2[1], scale=1, scale_units='xy', angles='xy', label='v2', color='green')

plt.quiver(0, 0, linear_combo[0], linear_combo[1],
           scale=1, scale_units='xy', angles='xy', color='red', label='a*v1 + b*v2')

plt.xlim(0, 10)
plt.ylim(0, 12)
plt.grid()
plt.legend()
plt.title("Linear Combination")
plt.show()
```

### Span
All possible linear combinations of those vectors.

#### Visualize span of one vector (a line)

```python
v = np.array([1, 2])

scalars = np.linspace(-5, 5, 20)
points = np.array([s * v for s in scalars])

plt.plot(points[:, 0], points[:, 1], '.')
plt.axhline(0)
plt.axvline(0)
plt.grid()
plt.title("Span of a Single Vector")
plt.show()
```

#### Visualize span of two vectors (plane)

```python
v1 = np.array([1, 0])
v2 = np.array([0, 1])

a_vals = np.linspace(-3, 3, 10)
b_vals = np.linspace(-3, 3, 10)

points = []
for a in a_vals:
    for b in b_vals:
        points.append(a * v1 + b * v2)

points = np.array(points)

plt.scatter(points[:, 0], points[:, 1], marker='.', color='red')
plt.grid()
plt.title("Span of v1 and v2 (Entire Plane)")
plt.show()
```

### Basis Vectors
The smallest set of vectors that can generate the entire space.

```python
e1 = np.array([1, 0])
e2 = np.array([0, 1])

v = np.array([4, -2])

# Represented using basis vectors
v = 4 * e1 + (-2) * e2

plt.quiver(0, 0, e1[0], e1[1], scale=1, scale_units='xy', angles='xy', label='e1', color='blue')
plt.quiver(0, 0, e2[0], e2[1], scale=1, scale_units='xy', angles='xy', label='e2', color='green')
plt.quiver(0, 0, v[0], v[1], scale=1, scale_units='xy', angles='xy', label='e2', color='red')

plt.xlim(-3, 5)
plt.ylim(-3, 5)
plt.grid()
plt.legend()
plt.title("Standard Basis Vectors")
plt.show()
```

### Matrix Multiplication = Composition of Transformations

```python
# Scaling then rotation

# Vector
v = np.array([1, 1])

# Scaling matrix (stretch x by 2)
M1 = np.array([
    [2, 0],
    [0, 1]
])

# Rotation 90 degrees CCW
M2 = np.array([
    [0, -1],
    [1, 0]
])

# Apply one by one
v1 = M1 @ v
v2 = M2 @ v1

# Combined matrix
print("v2:", v2)

plt.quiver(0, 0, v[0], v[1], scale_units='xy', angles='xy', color='g', scale=1, label='V1')
plt.quiver(0, 0, v1[0], v1[1], scale_units='xy', angles='xy', color='b', scale=1, label='V1 stretch x by 2')
plt.quiver(0, 0, v2[0], v2[1], scale_units='xy', angles='xy', color='r', scale=1, label='V1 Rotation 90 degrees')
plt.legend()
plt.xlim(-2, 3)
plt.ylim(-2, 3)
plt.grid()
plt.show()
```

### Associative Property

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[0, 1], [1, 0]])
C = np.array([[2, 0], [0, 2]])

left = (A @ B) @ C
right = A @ (B @ C)

print(left)
print(right)
print(np.allclose(left, right))  # True
```

### Matrix multiplication is NOT commutative
$A \cdot B \neq B \cdot A$

```python
M1 = np.array([[2, 0], [0, 1]])  # scale x
M2 = np.array([[0, -1], [1, 0]])  # rotate 90 deg

print("M2 @ M1:\n", M2 @ M1)
print("M1 @ M2:\n", M1 @ M2)
```

```python
# Visualizing
# Grid of points
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
points = np.array([[i, j] for i in x for j in y])

# Apply transformations
points1 = (M2 @ M1) @ points.T
points2 = (M1 @ M2) @ points.T

plt.scatter(points1[0], points1[1], label="M2 @ M1", alpha=0.5)
plt.scatter(points2[0], points2[1], label="M1 @ M2", alpha=0.5)
plt.grid()
plt.legend()
plt.title("Matrix Multiplication Order Matters")
plt.show()
```

### Three-dimensional linear transformations

```python
v = np.array([1, 1, 1])

matrices = {
    "identity": np.eye(3),
    "scale": np.diag([2, 1, 0.5]),
    "rotate_z": np.array([
        [0, -1, 0],
        [1, 0, 0],
        [0, 0, 1]
    ])
}

for name, A in matrices.items():
    print(name, "->", A @ v)
```

### Basis Vector Rule
Verify the columns rule

```python
A = np.array([
    [1, 2, 0],
    [0, 1, 1],
    [0, 0, 1]
])

e1 = np.array([1, 0, 0])
e2 = np.array([0, 1, 0])
e3 = np.array([0, 0, 1])

print(A @ e1)  # should equal first column
print(A @ e2)  # should equal second column
print(A @ e3)  # should equal third column
```

### Point Cloud Transformation
Transform many points

```python
# Generate random 3D points
points = np.random.randn(300, 3)

# Linear transformation matrix (shear-like)
A = np.array([
    [1, 0.5, 0],
    [0, 1, 0],
    [0, 0, 1]
])

# Apply transformation
points_t = points @ A.T

# Create figure
fig = plt.figure(figsize=(12, 5))

# Original points
ax1 = fig.add_subplot(121, projection='3d')
ax1.scatter(points[:, 0], points[:, 1], points[:, 2], color='red')
ax1.set_title("Original 3D Points")
ax1.set_xlabel("X")
ax1.set_ylabel("Y")
ax1.set_zlabel("Z")

# Transformed points
ax2 = fig.add_subplot(122, projection='3d')
ax2.scatter(points_t[:, 0], points_t[:, 1], points_t[:, 2], color='blue')
ax2.set_title("Transformed 3D Points")
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_zlabel("Z")

plt.tight_layout()
plt.show()
```

### Compose Transformations
Order matters

```python
scale = np.diag([2, 1, 1])
rotate = np.array([
    [0, -1, 0],
    [1, 0, 0],
    [0, 0, 1]
])

v = np.array([1, 1, 1])

print("scale  rotate:", rotate @ scale @ v)
print("rotate  scale:", scale @ rotate @ v)
```
