```python
import numpy as np
import matplotlib.pyplot as plt
```

### 2D Vectors

```python
v = np.array([3, 3])
print(v)

# Visualize
plt.quiver(0, 0, v[0], v[1], scale=1, scale_units='xy', angles='xy',color='g')
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.axhline(0)
plt.axvline(0)
plt.grid(True)
plt.title("2D Vector")
plt.show()
```

### 3D Vectors

```python
from mpl_toolkits.mplot3d import axes3d

v_3d = np.array([2,3,5])
print(v_3d)

# Visualize
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])

ax.quiver(0, 0, 0, v_3d[0], v_3d[1], v_3d[2])
ax.set_title("3D Vector")
# ax.view_init()
plt.show()
```

### Addition of 2D Vectors

```python
v1 = np.array([2, 1])
v2 = np.array([1, 3])
v_sum = v1 + v2

print(v_sum)

# Visualize
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, label='v1',color='blue')
plt.quiver(v1[0], v1[1], v2[0], v2[1], angles='xy', scale_units='xy', scale=1, label='v2',color='green')
plt.quiver(0, 0, v_sum[0], v_sum[1], angles='xy', scale_units='xy', scale=1, color='red', label='v1 + v2')

plt.xlim(0, 5)
plt.ylim(0, 5)
plt.grid()
plt.legend()
plt.title("Vector Addition")
plt.show()
```

### 2D Vector Scalar Multiplication (Scaling)

```python
v = np.array([2, 1])
scaled_v = 3 * v

print(scaled_v)

# Visualize
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, label='Original',color='green')
plt.quiver(0, 0, scaled_v[0], scaled_v[1], angles='xy', scale_units='xy', scale=1, color='red', label='Scaled')

plt.xlim(0, 8)
plt.ylim(0, 5)
plt.grid()
plt.legend()
plt.title("Vector Scaling")
plt.show()
```

### 3D Vector Addition

```python
v1 = np.array([2, 1, 3])
v2 = np.array([1, 3, 2])
v_sum = v1 + v2

print(v_sum)

# Visualize
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])

ax.quiver(0, 0, 0, v1[0], v1[1], v1[2],color='blue')
ax.quiver(v1[0], v1[1], v1[2], v2[0], v2[1], v2[2],color='green')
ax.quiver(0, 0, 0, v_sum[0], v_sum[1], v_sum[2], color='red')
ax.set_title("3D Vector Addition")
# ax.view_init()
plt.show()
```

### 3D Vector Scaling

```python
v = np.array([2, 1,3])
scaled_v = 2 * v

print(scaled_v)

# Visualize
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.set_xlim([0, 5])
ax.set_ylim([0, 5])
ax.set_zlim([0, 5])

ax.quiver(0, 0, 0, v[0], v[1], v[2],color='blue')
ax.quiver(0, 0, 0, scaled_v[0], scaled_v[1], scaled_v[2], color='red')
ax.set_title("3D Vector Scaling")
# ax.view_init()
plt.show()
```
