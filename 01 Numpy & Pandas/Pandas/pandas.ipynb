```python
import pandas as pd
```

### Read data from CSV file into dataframe

```python
df_csv = pd.read_csv("sales_data_sample.csv", encoding="latin1")
print(df_csv)
```

### Read data from JSON file into dataframe

```python
df_json = pd.read_json("sample_Data.json")
print(df_json)
```

### Read data from Excel file into dataframe

```python
df_excel = pd.read_excel("SampleSuperstore.xlsx")
print(df_excel)
```

### Creating Dataframe

```python
data = {
    "Name": ["Ram", "Shyam", "Ghanshyam", "Rupa", "Dev"],
    "Age": [10, 20, 22, 28, 25],
    "City": ['Mumbai', 'Toronto', 'Delhi', 'Delhi', 'Hyderabad']
}

df = pd.DataFrame(data)
print(df)
```

### Saving data into CSV, JSON and Excel file

```python
df.to_csv("CSVOutput.csv", index=False)
df.to_json("JSONOutput.json", index=False)
df.to_excel("ExcelOutput.xlsx", index=False)
```

- Means: Average of values in each column
- Std: Variation from mean value
- Min: Minimum value in each column
- 25%, 50%, 75%: After sorting, The median value or the value that divides 25%, 50%, 75% of dataset
- Max: Maximum value in each column
