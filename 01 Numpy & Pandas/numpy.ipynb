```python
import numpy as np
import time

from sympy.strategies.branch import condition
```

### Creating array from list

```python
arr_1d = np.array([1, 2, 3, 4, 5, 6])
print("1D Array: \n", arr_1d)

arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array: \n", arr_2d)
```

### List vs NumPy Array

```python
py_list = [1, 2, 3]
print("Python List Multiplication: ", py_list * 3)

np_arr = np.array([1, 2, 3])  # Element wise multiplication.
print("Numpy Array Multiplication: ", np_arr * 3)

# Efficiency
start = time.time()
py_list = [i * 3 for i in range(1000000)]
print("List Multiplication Operation time: ", time.time() - start)

start = time.time()
np_arr = np.arange(1, 1000000, 1) * 2
print("Numpy Multiplication time: ", time.time() - start)
```

### Creating array from scratch

```python
zeros = np.zeros([3, 4])
print("Zeros Array: \n", zeros)

ones = np.ones([3, 4])
print("Ones Array: \n", ones)

full = np.full([3, 4], 7)
print("Full Array: \n", full)

random = np.random.random([3, 4])  # Always between 0 and 1
print("Random Array: \n", random)

sequence = np.arange(1, 11, 2)
print("Sequence Array: \n", sequence)

randomInt = np.random.randint(1, 101, size=(3, 4))
print("Random Int Array: \n", randomInt)
```

### Vector, Matrices & Tensor

### Array Properties

```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])
print("Arr: \n", arr)
print("Shape: ", arr.shape)
print("Dimension: ", arr.ndim)
print("Size: ", arr.size)  # Number of Element
print("Data Type: ", arr.dtype)
```

### Array Reshaping

```python
arr = np.arange(1, 13)
print("Original Array: \n", arr)

reshaped = arr.reshape((3, 4))
print("Reshaped Array: \n", reshaped)

flattened = arr.flatten()
print("Flattened Array: \n", flattened)

raveled = reshaped.ravel()  # Return view, instead of copy
print("Raveled Array: \n", raveled)

transposed = reshaped.T
print("Transposed Array: \n", transposed)
```

### Array concatenation & stacking

```python
arr1 = np.arange(1, 11)
arr2 = np.arange(11, 21)

combined = arr1 + arr2  # It will add the element of both the array at a particular index
combined = np.concatenate((arr1, arr2))
print("Combined Array: \n", combined)

arr1 = np.array([[1, 2],
                 [3, 4]])
arr2 = np.array([[5, 6],
                 [7, 8]])
concatenated_array_row = np.concatenate((arr1, arr2), axis=0)  # axis 0 = row
print("Concatenated Array (Rows):\n", concatenated_array_row)
concatenated_array_column = np.concatenate((arr1, arr2), axis=1)  # axis 1 = column
print("Concatenated Array (Columns):\n", concatenated_array_column)

# Stack horizontally
stacked_array_h = np.hstack((arr1, arr2))
print("Stacked Horizontally:\n", stacked_array_h)

# Stack vertically
stacked_array_v = np.vstack((arr1, arr2))
print("Stacked Vertically:\n", stacked_array_v)

```

### NumPy Array Operations

#### Slicing & Negative indexing

```python
arr = np.arange(1, 11)
print("Original Array: \n", arr)
print("Basic Slicing: \n", arr[2:8])
print("Slicing with steps: \n", arr[2:8:2])
print("Negative Indexing: \n", arr[-5])
```

### Specific Element, Row & Column

```python
arr_2d = np.arange(1, 10).reshape((3, 3))
print("2D Array: \n", arr_2d)

print("Specific element: ", arr_2d[2, 1])
print("Entire row: \n", arr_2d[1])
print("Entire Column: \n", arr_2d[:, 2])
```

### Sorting

```python
unsorted_arr = np.random.randint(10, size=5)
print("Original Array: \n", unsorted_arr)
print("Sorted Array: \n", np.sort(unsorted_arr))

unsorted_arr_2d = np.random.randint(10, size=(3, 3))
print("Original 2D Array: \n", unsorted_arr_2d)
print("Sorted 2D Array (Column): \n", np.sort(unsorted_arr_2d, axis=0))  # Sorting column-wise
print("Sorted 2D Array (Row): \n", np.sort(unsorted_arr_2d, axis=1))  # Sorting row-wise
```

### Filtering

```python
numbers = np.arange(1, 11)
print("Numbers Array: \n", numbers)

even_numbers = numbers[numbers % 2 == 0]
print("Even Numbers: \n", even_numbers)
```

### Filtering with mask

```python
mask = numbers > 5
print("Numbers greater than 5: \n", numbers[mask])
```

### Fancy indexing vs np.where()

```python
indices = [0, 2, 4]
print(numbers[indices])

where_result = np.where(numbers % 3 == 0)
print("Where Result: \n", where_result)
print("NP Where Result: \n", numbers[where_result])
```

### Condition array using np.where()

```python
condition_array = np.where(numbers > 5, numbers * 4,
                           numbers - 4)  # Same as if numbers > 5: numbers *= 4 else: numbers -= 4
print("Condition Array: \n", condition_array)
```

### Compatibility

```python
arr1 = np.arange(1, 11)
arr2 = np.arange(11, 21)
arr3 = np.arange(21, 31)

print("Compatibility Shapes: ", arr1.shape == arr2.shape == arr3.shape)
```

### Vertical Stacking & Horizontal Stacking

```python
original = np.arange(1,10).reshape((3,3))
row = np.array([10,11,12])
col = np.array([[10],[11],[12]])

with_new_row = np.vstack((original, row))
print("Array with new row: \n", with_new_row)

with_new_column = np.hstack((original, col))
print("Array with new column: \n", with_new_column)
```
