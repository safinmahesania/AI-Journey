- Merging columns of two data frames

```python
# pd.merge(df1,df2, on="Common_Column_Name", how="Type of join")

# Inner Join
df_merged = pd.merge(df_1, df_2, on="customer_id", how="inner")
print("Inner Join")
print(df_merged)
```

```python
# Right Join
df_merged = pd.merge(df_1, df_2, on="customer_id", how="right")
print("Right Join")
print(df_merged)
```
